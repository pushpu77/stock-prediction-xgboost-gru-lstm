```python
# Noise Detection


Noisy_data=pd.DataFrame()

for i in df.columns:
    if i == "Open":
        outlier=df[(df['Open']>160) & (df['Adj Close']<180) ]       # According to Scatter plot
        Noisy_data = Noisy_data._append(outlier)
    if i == "High":
        outlier=df[(df['High']>160) & (df['Adj Close']<180) ]       # According to Scatter plot
        Noisy_data = Noisy_data._append(outlier)
    if i == "Low":
        outlier=df[(df['Low']>160) & (df['Adj Close']<180) ]        # According to Scatter plot
        Noisy_data = Noisy_data._append(outlier)
    if i == "Close":
        outlier=df[(df['Close']>160) & (df['Adj Close']<180) ]      # According to Scatter plot
        Noisy_data = Noisy_data._append(outlier)


Noisy_data
```

```python
#Price_Date_Volume chart


from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Each color can be define by RGB for example black is (0,0,0)



# make secondary plot in yaxis

fig2 = make_subplots(specs=[[{"secondary_y": True}]])

# first plot
fig2.add_trace(go.Scatter(x=df1004["Date"],y=df1004['Adj Close'],name='Price',mode="lines",line= dict(color='rgb(100,120,180)', width=2)),secondary_y=False  )


# secondary plot
fig2.add_trace(go.Bar(x=df1004["Date"],y=df1004['Volume'],name='Volume',marker= dict(color='rgb(80,80,0)')),secondary_y=True )




fig2.update_yaxes(range=[0,30000000000],secondary_y=True,  showgrid=False)
#fig2.update_yaxes(visible=True, secondary_y=True)




fig2.update_layout(height=600, width=1000,
                  title_text="Time price chart in yearly time frame (price and trading volume)")



# General setting
fig2.update_layout( xaxis_title='Date',
    xaxis=dict(
        showline=True , showgrid=False ,showticklabels=True ,
        linecolor= 'rgb(204,204,204)',
        linewidth=2, ticks='outside', tickfont=dict(
            family="Arial", size=12 , color='rgb(82,82,82)',
        ),
    ), yaxis_title='Price',
                  yaxis=dict(showline=True , showgrid=False ,showticklabels=True ,
        linecolor= 'rgb(204,204,204)',
        linewidth=2, ticks='outside', tickfont=dict(
            family="Arial", size=12 , color='rgb(82,82,82)',
        ),),
    autosize=False,
                  margin=dict(
                      autoexpand=False , l=100 , r= 20 , t= 110 ,) ,
     plot_bgcolor="white")



fig2.update_yaxes(title_text="Volume", secondary_y=True)



fig2.show()
```

```python
# Force lowercase (optional)
df1008.columns = [x.lower() for x in df1008.columns]


# Construct a 2 x 1 Plotly figure
fig = make_subplots(rows=2, cols=1)


# price Line
fig.append_trace(
    go.Scatter(
        x=df1008['date'],
        y=df1008['open'],
        line=dict(color='rgb(0,100,200)', width=1),
        name='Open',
        # showlegend=False,
        legendgroup='1',
    ), row=1, col=1
)


# Candlestick chart for pricing
fig.append_trace(
    go.Candlestick(
        x=df1008['date'],
        open=df1008['open'],
        high=df1008['high'],
        low=df1008['low'],
        close=df1008['close'],
        increasing_line_color='rgb(0,177,64)',
        decreasing_line_color='rgb(255,0,0)',
        showlegend=False
    ), row=1, col=1
)


# Fast Signal (%k)
fig.append_trace(
    go.Scatter(
        x=df1008['date'],
        y=df1008['macd_12_26_9'],
        line=dict(color='rgb(0,0,0)', width=1),
        name='MACD',
        # showlegend=False,
        legendgroup='2',
    ), row=2, col=1
)


# Slow signal (%d)
fig.append_trace(
    go.Scatter(
        x=df1008['date'],
        y=df1008['macds_12_26_9'],
        line=dict(color='rgb(0,181,226)', width=1),
        # showlegend=False,
        legendgroup='2',
        name='Signal'
    ), row=2, col=1
)


# Colorize the histogram values
colors = np.where(df1008['macdh_12_26_9'] < 0, 'rgb(0,177,64)', 'rgb(255,0,0)')


# Plot the histogram
fig.append_trace(
    go.Bar(
        x=df1008['date'],
        y=df1008['macdh_12_26_9'],
        name='Histogram',
        marker_color=colors,
    ), row=2, col=1
)


# Make it pretty
layout = go.Layout(
    plot_bgcolor='rgb(255,255,255)',
    # Font Families
    font_family='Arial',
    font_color='rgb(82,82,82)',
    font_size=12,
    xaxis=dict(
        rangeslider=dict(
            visible=False
        )
    )
)


# Update options and show plot
fig.update_layout(layout)


fig.show()
```
